```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
final_df = pd.concat([df, dummies], axis='columns')
final_df
```

```python
final_df = final_df.drop(['town', 'west windsor'], axis='columns')
final_df
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

```python
X = final_df.drop(['price'], axis='columns')
X
```

```python
y = final_df.price
y
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[2800,0, 1]])
```

```python
model.predict([[3400,0, 0]])
```

# One Hot Encoding using Sklearn


```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = df[['town', 'area']].values
X
```

```python
y = df.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])],
    remainder='passthrough' 
)
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X, y)
```

```python
model.predict([[1,0,2800]])
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'])  # Matplotlib plot.
plt.show()
```

```python
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
final_df = pd.concat([df, dummies], axis='columns')
final_df
```

```python
final_df = final_df.drop(['Car Model','Mercedez Benz C class'], axis='columns')
final_df
```

```python
X = final_df.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
final_df
```

```python
#Price of mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000, 4, 0, 0]])
```

```python
#Price of BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```

```python
model.score(X, y)
```
