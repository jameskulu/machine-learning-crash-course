# Dropdown selector and Slider

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
app = dash.Dash()
```

```python
app.layout = html.Div([
    html.Label('Choose: '),
    dcc.Dropdown(
        id = "first-dropdown",
        options = [
            {'label':'Name','value':'Name'},
            {'label':'Age','value':'Age'},
            {'label':'Percetage','value':'Percentage'}
        ],
        value = 'Name',
        multi = True
    ),
    html.Label("Slider"),

#     SLider
    dcc.Slider(
        min = 1,
        max = 10,
        value = 5,
        step = .5,
        marks = {i:i for i in range(10)}
    ),
    
#     Range Slider
     html.Label("Range Slider"),
    dcc.RangeSlider(
        min = 1,
        max = 10,
        step = 5,
        value = [3,7],
        marks = {i:i for i in range(10)}
    ),
    
    
#     Input Field
     html.Label("Input Field"),
    dcc.Input(
        placeholder = "Enter your name",
        value = '',
        type = 'text'
    )
])
```

```python
if __name__ == '__main__':
    app.run_server()
```
